```python
# 设立爬取关键字
attitude_keyurl = r'<a href="https://weibo.cn/attitude/.*?</a>'
repost_keyurl = r'<a href="https://weibo.cn/repost/.*?</a>'
comment_keyurl = r'<a href="https://weibo.cn/comment/.{45,52}>评论.*?</a>'
###设定weiboid检索关键词
def makeattitudeurllist(x):
    uum = weibodf.iloc[x]
    attitude_url = ['https://weibo.cn/attitude/'+uum['id']+'?rl=1&page='+str(x) 
                    for x in range(1,int(uum['attitude_num'])//10+2,1)]
    return(attitude_url)
def makereposturllist(x):
    uum = weibodf.iloc[x]
    repost_url = ['https://weibo.cn/repost/'+uum['id']+'?uid='+uum['uid']+'&&page='+str(x) 
                  for x in range(1,int(uum['repost_num'])//10+2,1)]
    return(repost_url)
def makecommenturllist(x):
    uum = weibodf.iloc[x]
    comment_url = ['https://weibo.cn/comment/'+uum['id']+'?&uid='+uum['uid']+'&&page='+str(x) 
                   for x in range(1,int(uum['comment_num'])//10+2,1)]
    return(comment_url)
```

```python
df=pd.read_excel(r'C:\Users\yingshan.Zhang\Desktop\爬虫\Task_df.xlsx',index=False,encoding='gb18030')
df.columns=['task_name','taskurl']
df=df.iloc[234:237]
for i,v in zip(df['taskurl'],df['task_name']):
    kurl=i
    driver.get(i)
    weibodf = pd.DataFrame()
    uidlist = pd.Series()
    pagenum_com = r'"跳页" />.*?页'
    page=re.findall(pagenum_com,driver.page_source)[0]
    if page[-1]=='页':
        pagenum = int(re.findall(pagenum_com,driver.page_source)[0][10:12])
    else:
        pagenum = int(re.findall(pagenum_com,driver.page_source)[0][10:11])
    for m in range(1,pagenum+1,1):
        url=i+str(m)
        driver.get(url)
        time.sleep(1.8)
        df = pd.DataFrame({
            'attitude':re.findall(attitude_keyurl,driver.page_source),
            'repost':re.findall(repost_keyurl,driver.page_source),
            'comment':re.findall(comment_keyurl,driver.page_source)})
        weibodf = pd.concat([weibodf,df])
    uid = pd.Series(re.findall(r'uid=[0-9]{8,10}&',driver.page_source)).map(lambda x:x[4:-1])
    uidlist = pd.concat([uidlist,uid])
    weibodf['No']=range(len(weibodf))
    weibodf['id'] = weibodf['repost'].map(lambda x:re.findall('[A-Za-z][A-Z0-9a-z]{8}',x)).map(lambda x:x[0])
    weibodf['uid'] = weibodf['repost'].map(lambda x:re.findall('[0-9]{8,10}',x)).map(lambda x:x[0])
    weibodf['attitude_num'] = weibodf['attitude'].map(lambda x:re.findall(r'赞.*?<',x)).map(lambda x:x[0]).map(lambda x:x[2:-2]).map(lambda x:int(x))
    weibodf['comment_num'] = weibodf['comment'].map(lambda x:re.findall(r'评论.*?<',x)).map(lambda x:x[0]).map(lambda x:x[3:-2]).map(lambda x:int(x))
    weibodf['attitude_list'] = weibodf['No'].map(makeattitudeurllist)
    weibodf['comment_list'] = weibodf['No'].map(makecommenturllist)
    weibodf['all_list'] = weibodf['attitude_list']+weibodf['comment_list']
    weibodf['total_num']=weibodf['attitude_num'].map(int).sum()+weibodf['comment_num'].map(int).sum()
    total_num=weibodf['attitude_num'].map(int).sum()+weibodf['comment_num'].map(int).sum()
    list=[]
    for i in weibodf['all_list']:
        for m in i:
            list.append(m)
    print('任务名称:'+v+'-----'+'微博ID任务数量:'+str(total_num)+'-----'+'任务列表数量:'+str(len(list)))
    ma=0
    for j in list[ma:]:
        driver.get(j)
        sys.stdout.write('百分比: {:.2%}'.format(len(uidlist)/total_num)+'---'+'已爬取数量:'+str(len(uidlist))+'\r')
        sys.stdout.flush()
        seconds=random.uniform(1.4,1.7)
        time.sleep(seconds)
        uid = pd.Series(re.findall(r'/u/[0-9]{8,10}',driver.page_source)).map(lambda x:x[3:])
        uidlist = pd.concat([uidlist,uid])
        uiddr=uidlist.drop_duplicates()
        ma +=1
    uidlist.to_csv(str(ma*10)+'---'+str(len(uidlist))+'---'+str(len(uiddr))+'---'+v+'03uid.csv',index=False,encoding='gb18030')
```
