# 1评论正文数据格式

```python
pinglun = pd.read_csv('1229/df21229.csv')
pinglun.head(5)
```

# 2词包数据格式

```python
cibao = pd.read_excel('1229/cibao1229.xlsx')
cibao
```

# 3。1 词包检索

```python
df = pinglun.copy()
for i in cibao.检索词:
    df = keyword_columns(df,i)
df.head(5)
```

# 3。2 词频统计

```python
cdf = df.sum(axis=0)
cdf[1:]
```

# 3。3 结果展示

```python
cibao['频数'] = cibao['检索词'].map(cdf)
res = cibao.groupby(['正负向','一级属性'])[['频数']].sum()
res['占比'] = (res['频数']/res['频数'].sum()).map(lambda x: str(100*x)[:4] +'%')
res.to_excel('res2.xlsx')
res
```
