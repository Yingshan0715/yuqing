```python
df = pinglun.copy()
for i in cibao.关键词:
    df = keyword_columns(df,i)
df.head(5)
```

```python
cdf = df.sum(axis=0)
cdf[1:]
```

```python
cibao['频数'] = cibao['检索词'].map(cdf)
res = cibao.groupby(['正负向','一级属性','检索词'])[['频数']].sum()
res['占比'] = (res['频数']/res['频数'].sum()).map(lambda x: str(100*x)[:4] +'%')
res.to_excel('兰蔻菁纯柔润丝缎唇膏口红持久显色滋润保湿赤茶橘120王俊凯.xlsx')
res
```
