```python
df=pd.read_excel(r'C:\Users\yingshan.Zhang\Desktop\爬虫\task_dir.xlsx',index=False,encoding='gb18030')
df.columns=['task_name','taskurl']
df['id']=[i[28:] for i in df['taskurl']]
df['uid']=[i[28:37] for i in df['taskurl']]
df['NO']=range(len(df))
ws=176
we=177
for i,v,k,m,a in zip(df.iloc[ws:we]['taskurl'],df.iloc[ws:we]['uid'],df.iloc[ws:we]['id'],df.iloc[ws:we]['task_name'],df.iloc[ws:we]['NO']):
    weibodf = pd.DataFrame()
    uidlist = pd.Series()
    driver.get(i)
    uum=df.iloc[a]
    try:
        if len(re.findall('#attitude">赞(.*?)</a>\xa0</span><br',driver.page_source)[0][1:-1])!=0:
            attitude_page_num=int(re.findall('#attitude">赞(.*?)</a>\xa0</span><br',driver.page_source)[0][1:-1])//10+1
            attitude_list=["https://weibo.cn/attitude/"+str(uum['uid'])+"?&page="+str(x) for x in range(1,attitude_page_num+1)]
        else:
            pass
    except Exception as err:
        pass
    try:
        if len(re.findall('#rt">转发(.*?)</a>\xa0</span>',driver.page_source)[0][1:-1])!=0:
            repost_page_num=int(re.findall('#rt">转发(.*?)</a>\xa0</span>',driver.page_source)[0][1:-1])//10+1
            repost_list=["https://weibo.cn/repost/"+str(uum['uid'])+"?&page="+str(x) for x in range(1,repost_page_num+1)]
        else:
            pass
    except Exception as err:
        pass
    try:
        if len(re.findall('<span class="pms">\xa0评论(.*?)\xa0</span>',driver.page_source)[0][1:-1])!=0:
            comment_page_num=int(re.findall('<span class="pms">\xa0评论(.*?)\xa0</span>',driver.page_source)[0][1:-1])//10+1
            comment_list=["https://weibo.cn/comment/"+str(uum['uid'])+"?&page="+str(x) for x in range(1,comment_page_num+1)]
        else:
            pass
    except Exception as err:
        pass
    all_list=attitude_list+comment_list
    print(m+'-----'+'任务列表数量:'+str(len(all_list)))
    ma=0
    for j in all_list:
        driver.get(j)
        driver.set_window_size(300,600)
        sys.stdout.write('已爬取数量:'+str(len(uidlist))+'\r')
        sys.stdout.flush()
        seconds=random.uniform(1.5,2.5)
        time.sleep(seconds)
        uid = pd.Series(re.findall(r'/u/[0-9]{8,10}',driver.page_source)).map(lambda x:x[3:])
        uidlist = pd.concat([uidlist,uid])
        uiddr=uidlist.drop_duplicates()
        ma += 1
    uidlist.to_csv(str(ma*10)+'---'+str(len(uidlist))+'---'+str(len(uiddr))+'---'+m+'uid.csv',index=False,encoding='gb18030')
```
